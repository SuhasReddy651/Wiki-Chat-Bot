```python
import sqlite3
import pandas as pd
```

```python
DB_PATH = "chatbot.db"
```

```python
def fetch_chat_messages():
    """Fetch all rows from the chat_messages table."""
    try:
        # Connect to the database
        conn = sqlite3.connect(DB_PATH)
        # Query to fetch all rows from chat_messages
        query = "SELECT * FROM chat_messages"
        # Read the query into a pandas DataFrame
        df = pd.read_sql_query(query, conn)
        return df
    except Exception as e:
        print(f"Error accessing the database: {e}")
        return None
    finally:
        if conn:
            conn.close()
```

```python
# Call the function to fetch data
chat_messages_df = fetch_chat_messages()

# Display the data
if chat_messages_df is not None:
    display(chat_messages_df)  # Use `display` in Jupyter for better output
else:
    print("No data available or failed to fetch.")
```
